```python
def addFeatures(data):
    # wine
    wines=["Red wine","Sparkling wine & champagne","White wine","Rosé wine"]

    data["WINE"]=False
    for category in wines:
        data.loc[data["CATEGORY"] == category,"WINE"]=True
    data["WINE"].describe()
    only_wines=data[data["WINE"]]
    
    # red wine, white wine, sparkling etc
    data["RED WINE"]=False
    data.loc[data["CATEGORY"]==wines[0],"RED WINE"]=True
    data["WHITE WINE"]=False
    data.loc[data["CATEGORY"]==wines[2],"WHITE WINE"]=True
    data["SPARKLING"]=False
    data.loc[data["CATEGORY"]==wines[1],"SPARKLING"]=True
    data["ROSE"]=False
    data.loc[data["CATEGORY"]==wines[3],"ROSE"]=True
    return data
data2=addFeatures(data)
```

```python
import re
def cleanPriceColumn(column):
    substitute = lambda string: re.sub("\sEUR", "",string)
    substitute_space = lambda string: re.sub("\s+", "",string)
    cleaned_column=column.apply(substitute)
    cleaned_column=cleaned_column.apply(substitute_space)
    cleaned_column=cleaned_column.apply(float)
    return cleaned_column

data["PRICE"]=cleanPriceColumn(data["PRICE / LITRE"])

#substitute = lambda string: re.sub("\sEUR", "",string)
#substitute_space = lambda string: re.sub("\s+", "",string)
#data["PRICE"]=data["PRICE / LITRE"].apply(substitute)
#data["PRICE"]=data["PRICE"].apply(substitute_space)
#data["PRICE"]=data["PRICE"].apply(float)
```

```python
# Clean Acids
def acidsToNumerical(data):
    substitute=lambda string: re.sub("g/l","",str(string))
    data=data.apply(substitute)
    data=data.apply(float)
    return data
data["ACIDS"]=acidsToNumerical(data["ACIDS"])
```

```python
# Clean alcohol field
def cleanAlcohol(data):
    substitute=lambda string: re.sub("%","",string)
    data=data.apply(substitute)
    data=data.apply(float)
    return data
data["ALCOHOL"]=cleanAlcohol(data["ALCOHOL"])
```

```python
data.keys()
```

```python
# clean sugar field
def cleanSugar(data):
    substitute = lambda string: re.sub("g/l","",str(string))
    data=data.apply(substitute)
    data=data.apply(float)
    return data
data["SUGAR"]=cleanSugar(data["SUGAR"])
```

```python
# clean energy field
def cleanEnergy(data):
    substitute = lambda string: re.sub("Kcal / 100ml","",str(string))
    data=data.apply(substitute)
    data=data.apply(float)
    return data
data["ENERGY"]=cleanEnergy(data["ENERGY"])
```

```python

def extractVintage(string):
    vintage=re.match(".*([1-3][0-9]{3})",string)
    try:
        vintage=int(vintage.group(1))
    except:
        pass
    return vintage
try_this=data["NAME"][4]
extractVintage(try_this)
data["VINTAGE"]=data["NAME"].apply(extractVintage)
```

```python
index=data["WINE"] & (data["VINTAGE"]>1985)
sns.boxplot(x="VINTAGE", y="PRICE",data=data.loc[index])
plt.xticks(rotation=90)
plt.ylim(0,300)
#plt.xlim(1990,2018)
plt.title("Price per litre for each vintage since 1988")
plt.show()
```
